# Q1) Create a Doc object from the file owlcreek.txt

```python
import spacy
nlp = spacy.load('en_core_web_sm')
with open ("owlcreek.txt", "r") as file:
    text = file.read()
    doc = nlp(text)
```

# Q2) How many tokens are contained in the file?

```python
 for t in doc:
    print(t)
```

# Q3) How many sentences are contained in the file?

```python
sent = list(doc.sents)
print(f"len of sentences: {len(sent)}")
```

# Q6) Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text.

```python
from spacy.matcher import PhraseMatcher 

phrase_matcher = PhraseMatcher(nlp.vocab)

phrases = ["swimming\nvigorously"]
patterns = [nlp(text) for text in phrases]
phrase_matcher.add("Pattern",patterns)

doc = nlp(text)
matches = phrase_matcher(doc)
matches
```
